implying the error factor. The likelihood of an intruder in the passenger population is P(I = 1) =
0.00001. What are the chances that an alarm would be triggered when an individual is actually an
intruder?
Solution:

We need to compute the posterior probability of an intruder given that the alarm has been triggered. We can use Bayes' theorem to compute this probability:

P(I = 1|A = 1) = P(A = 1|I = 1) * P(I = 1) / P(A = 1)

We know that P(A = 1|I = 1) = 0.98, P(A = 1|I = 0) = 0.001, and P(I = 1) = 0.00001. To compute P(A = 1), we can use the law of total probability:

P(A = 1) = P(A = 1|I = 1) * P(I = 1) + P(A = 1|I = 0) * P(I = 0)

We know that P(I = 0) = 1 - P(I = 1) = 0.99999. Substituting these values, we get:

P(A = 1) = 0.98 * 0.00001 + 0.001 * 0.99999 = 0.00109

Now we can compute the posterior probability of an intruder given that the alarm has been triggered:

P(I = 1|A = 1) = 0.98 * 0.00001 / 0.00109 = 0.00899

So the chances that an alarm would be triggered when an individual is actually an intruder are about 0.9%.

= 1 - (0.95)^6
= 0.2642

10 * 0.2642 = 2.642 customers (rounded to 3 customers)

So on a daily basis, there are approximately 3 customers that come into the bank.
